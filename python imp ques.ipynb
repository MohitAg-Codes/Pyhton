```python
n = input()
ch = ""

for i in n:
    if i not in ch:
        ch += i

print(len(ch))
```

```python
n = 5
for i in range(n):
    for j in range(i+1,n):
        print(' ',end = '')
    for j in range(i+1):
        print('*',end='')
    print()
```

```python
n = 5
for i in range(n):
    for j in range(i+1,n):
        print(' ',end=' ')
    for j in range(i+1):
        print('*',end = ' ')
    for j in range(i):
        print('*',end=' ')
    print()
```

```python
n =5
for i in range(1,n+1):
    for j in range(i,n):
        print(' ',end='')
    for j in range(i):
        print(i,end = ' ')
    print()

for i in range(n-1,0,-1):
    for j in range(i,n):
        print(' ',end='')
    for j in range(i):
        print(i, end = ' ')
    print()
```
