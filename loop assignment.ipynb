```python
prices = {"Laptop": 1000, "Mouse": 50, "Monitor": 300}

orders = [
    (101, "Laptop", 1, "INTL"),
    (102, "Mouse", 1, "LOCAL"),
    (103, "Monitor", 1, "OTHER")
]


final_cost = {}
total_revenue = 0

for order_id, product, quantity, destination in orders:
    cost = prices[product] * quantity

    if destination == "INTL":
        cost += cost * 0.15
    elif destination == "LOCAL":
        cost += cost * 0.05
    # else: no tax

    final_cost[order_id] = cost
    total_revenue += cost

print(final_cost)
print("Total Revenue:", total_revenue)
```

```python
portfolio = [("AAPL", 50), ("GOOG", 10), ("TSLA", 5)]

target_allocation = {"AAPL": 0.4, "GOOG": 0.3, "TSLA": 0.3}

total_portfolio_value = 100000

Static prices (helper function ki jagah)
prices = {
    "AAPL": 200,
    "GOOG": 3000,
    "TSLA": 1000
}


for ticker, shares in portfolio:
    current_value = prices[ticker] * shares
    current_allocation = current_value / total_portfolio_value
    target = target_allocation[ticker]

    if current_allocation > target:
        print(ticker, ": Sell")
    elif current_allocation < target:
        print(ticker, ": Buy")
    else:
        print(ticker, ": Hold")
```

```python
cart = [("User1", "Book", 2), ("User1", "Pen", 5), ("User2", "Eraser", 1)]

discounts = {"VIP": 0.20, "Regular": 0.05}

user_status = {"User1": "VIP", "User2": "Regular"}

prices = {"Book": 10, "Pen": 2, "Eraser": 1}

final_bill = {}

for user, item, qty in cart:
    cost = prices[item] * qty

    if cost > 50:
        status = user_status[user]
        discount = discounts[status]
        cost -= cost * discount   # apply discount

    final_bill[user] = final_bill.get(user, 0) + cost

print(final_bill)
```

```python
timesheet = [("Emp001", 45), ("Emp002", 30), ("Emp003", 60)]
rates = {"Emp001": 20, "Emp002": 25, "Emp003": 15}

weekly_pay = []
for empid , hours in timesheet:
    rate = rates[empid]
    
    if hours > 50:
        pay = 40 * rate
        pay += 10 * rate * 1.5
        pay += (hours - 50) * rate * 2
    elif hours > 40:
        pay = 40 * rate
        pay += (hours - 40) * rate * 1.5
    else:
        pay = hours * rate

    weekly_pay.append((empid,pay))
print(weekly_pay)   
    
```

```python
sensors = [("SensorA", 100, "F"), ("SensorB", 25, "C"), ("SensorC", 300, "K")]
limits = {"SensorA": 35, "SensorB": 30, "SensorC": 30} 

warnings = []
for name,temp,unit in sensors:
    if unit == 'F':
        temp_c = (temp-32) * 5/9
    elif unit == 'K':
        temp_c = temp - 273.15
    else:
        temp_c = temp

    if temp>limits[name]:
        excess = round(temp_c - limits[name],2)
        warnings.append((name,excess))

print(warnings)
    
        
```

```python
actions = [("Hero", "Attack", 20), ("Hero", "Heal", 10), ("Enemy", "Attack", 15)]
stats = {"Hero_HP": 100, "Enemy_HP": 50}

for actor, action, value in actions:
    if action == 'Attack':
        if actor == 'Hero':
            stats["Enemy_HP"] -= value
        else:
            stats["Hero_HP"] -= value

    elif action == 'Heal':
        key = actor + "_HP"
        stats[key] = min(100,stats[key] + value)
        
    else:
        print("Missed turn")

print(stats)
```

```python
backpack = [("Wood", 5), ("Stone", 2), ("Iron", 0)]
recipes = {"Sword": {"Wood": 1, "Iron": 2}, "Shield": {"Wood": 3, "Stone": 1}}

backpack_dict = dict(backpack)
craftable_items = []
shopping_list = {}

for item, ingredients in recipes.items():
    can_craft = True

    for material, required_qty in ingredients.items():
        available_qty = backpack_dict[material]

        if available_qty < required_qty:
            can_craft = False
            shopping_list[material] = required_qty - available_qty

    if can_craft:
        craftable_items.append(item)


print("Craftable Items:", craftable_items)
print("Shopping List:", shopping_list)
```

```python
racers = [("CarA", 120), ("CarB", 115), ("CarC", 130)] 
penalties = {"CarA": 5, "CarC": 0} 

final_times = {}

for name,time in racers:
    if name in penalties:
        final_times[name] = time + penalties[name]
    else:
        final_times[name] = time
        
winner = min(final_times, key = final_times.get)
    
print("Final Times:", final_times)
print("Winner:", winner)
```

```python
moves = [("N", 1), ("E", 2), ("S", 1)]
position = {"x": 0, "y": 0}
boundaries = (-5, 5)

for direction, steps in moves:
    prev_y = position['y']
    prev_x = position['x']
    
    if direction == 'N':
        position['y'] += steps
    elif direction == 'E':
        position['x'] += steps
    else:
        position["y"] -= steps

    if not (boundaries[0] <= position['y'] <= boundaries[1] and
            boundaries[0] <= position['x'] <= boundaries[1]):
        position['y'] = prev_y
        position['x'] = prev_x

print(position)
```

```python
passengers = [("John", "A1"), ("Doe", "B2"), ("Smith", "A1")]
seat_map = {"A1": "Taken", "B2": "Empty", "A2": "Empty"}

confirmed = []
standby = []

for name, preferred_seat in passengers:

    # Case 1: Preferred seat is empty
    if seat_map.get(preferred_seat) == "Empty":
        seat_map[preferred_seat] = "Taken"
        confirmed.append((name, preferred_seat))

    # Case 2: Preferred seat is taken → assign first available seat
    elif seat_map.get(preferred_seat) == "Taken":
        assigned = False

        for seat, status in seat_map.items():
            if status == "Empty":
                seat_map[seat] = "Taken"
                confirmed.append((name, seat))
                assigned = True
                break

        if not assigned:
            standby.append(name)

    # Case 3: Seat does not exist
    else:
        standby.append(name)

print("Confirmed:", confirmed)
print("Standby:", standby)
print("Final Seat Map:", seat_map)
               
    
```

```python
requests = [("Alice", 2), ("Bob", 1), ("Charlie", 5)]
venue_capacity = 100
vip_list = {"Alice": True, "Bob": False}

guest = {}

for name ,size in requests:
    if name in vip_list:
        guest[name] = size

    elif venue_capacity - size > 0:
        guest[name] = size
        venue_capacity -= size

print(guest)
print("Remaining Capacity:", venue_capacity)
```

```python
emails = [("john@gmail.com", "User"), ("admin@site.com", "Admin"), ("fake.com", "Bot")]
banned_domains = {"yahoo.com": "Deprecated", "hotmail.com": "Deprecated"}

invalid = []
flagged = []
valid_users = {}

for x,role in emails:
    if '@' not in x:
        invalid.append(x)
        continue

    user, domain = x.split('@')
    
    if domain in banned_domains:
        flagged.append(x)
    else:
        valid_users[x] = role

print("Invalid:", invalid)
print("Flagged:", flagged)
print("Valid Users:", valid_users)   
```

```python
samples = [("Seq1", "ATCG"), ("Seq2", "ATGG"), ("Seq3", "TTTT")]
reference = {"Seq1": "ATCG", "Seq2": "ATCG", "Seq3": "ATCG"}

mutation_status = {}

for x,y in samples:
    z = reference[x]
    if y == z:
        status = 'Clean'
    elif len(y) != len(z):
        status = 'Frameshift'
    else:
        status = 'Point Mutation'

    mutation_status[x] = status

print(mutation_status)
```

```python
inputs = [("I want a refund", 5), ("Hello there", 1), ("Cancel my sub", 4)] 
keywords = {"refund": "Support", "cancel": "Retention", "hello": "Greeting"}

route_list = []

for text,priority in inputs:
    text_lower = text.lower()
    found_keyword = False

    for keyword in keywords:
        if keyword in text_lower:
            found_keyword = True
            break

    if priority > 3 and found_keyword:
        route = "Human Agent"
    elif found_keyword:
        route = "Bot"
    else:
        route = "General Menu"

    route_list.append((text,route))

print(route_list)
```

```python
classes = [("Math", "Mon", 9), ("History", "Mon", 9), ("Art", "Tue", 10)]
room_availability = {"Mon": [9, 10, 11], "Tue": [9, 10]}

approved = []  
conflicts = []

for classname, day, time in classes:
    if time in room_availability[day]:
        approved.append((classname, day, time))
        room_availability[day].remove(time)

    else:
        conflicts.append((classname, day, time))

print("Approved Classes:", approved)
print("Conflicting Classes:", conflicts)
        
```

```python
trades = [("BTC", "Buy", 0.5), ("ETH", "Sell", 10)]
wallets = {"BTC": 1.0, "ETH": 5.0}
fees = {"Buy": 0.01, "Sell": 0.02}

cash = 0

for coin, trade_type, amount in trades:

    if trade_type == 'Buy':
        wallets[coin] = wallets[coin] + amount * (1 - fees[trade_type])
    elif trade_type == 'Sell':
        if wallets[coin] > amount:
            wallets[coin] -= amount
            cash = cash + amount*(1- fees[trade_type])
        else:
            print('insuffient fund')

    else:
        print('invalid Type')

print(wallets)
print("Cash Balance:",cash)
    
    
```
