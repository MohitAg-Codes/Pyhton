9. Write a function validate_age that takes an integer age. If age is less than 0, return "Invalid".
If it is under 18, return "Minor". Otherwise, return "Adult"

```python
def validate_age(age):
    if age < 0:
        return "Invalid"
    elif age < 18:
        return "Minor"
    else:
        return "Adult"


print(validate_age(-1))   # Invalid
print(validate_age(10))   # Minor
print(validate_age(25))   # Adult
```

12. Write a function find_max that accepts a list of integers and returns the largest number in
that list (do not use the built-in max() function; implement your own logic).

```python
def max(n):
    max_num = float('-inf')
    for i in n:
        if i > max_num:
            max_num = i
    return max_num

a = [10,20,14,50,99,23,56]
print(max(a))
```

13. Write a function filter_even that takes a list of integers and returns a new list containing
only the even numbers.

```python
def filter_even(n):
    new = []
    for i in n:
        if i % 2== 0:
            new.append(i)
    return new

a = [10,20,14,50,99,23,56]
print(filter_even(a))
```

5. Write a function get_first_last that accepts a list and returns a tuple containing just the first
and the last element of that list.

```python
def get_first_last(lst):
    if not lst:
        return None
    else:
        return (lst[0], lst[-1])
  

print(get_first_last([1, 2, 3, 4, 5]))  
print(get_first_last(['a', 'b', 'c']))   
print(get_first_last([]))                
```

18. Write a function count_characters that takes a string and returns a dictionary where keys
are characters and values are the number of times they appear in the string.


```python
def count_characters(n):
    _dict = {}
    for i in n:
        if i in _dict:
            _dict[i] += 1
        else:
            _dict[i] = 1
    return _dict

print(count_characters("hello world"))
```

19. Write a function update_inventory that takes an inventory dictionary {'item': quantity}, an
item_name, and an amount_to_add. Update the dictionary and return the new quantity for
that item

```python
def update_inventory(inventory, item_name, amount):
    if item_name in inventory:
        inventory[item_name] += amount
    else:
        inventory[item_name] = amount
    return inventory

inventory = {'apple':10 , 'bnana' : 5}
print(update_inventory(inventory, 'apple', 3))  
print(update_inventory(inventory, 'orange', 7))
print(inventory) 
```

21. Write a function find_target that takes a list of numbers and a target number. Use a for loop
to check each number. If the target is found, print "Found it!" and immediately use break to
stop the loop.


```python
def find_target(n,target):
    for i in n:
        if i == target:
            return 'Found it'
    return 'not found'

n = [1,2,3,4,5]
print(find_target(n,1))
print(find_target(n,7))
```

23. Write a function sum_until_negative that takes a list of numbers. Use a loop to sum the
numbers, but if you encounter a negative number, use break to stop the loop immediately
and return the sum calculated so far

```python
def sum_until_negative(n):
    total = 0
    for i in n:
        if i < 0:
            break
        total += i
    return total

print(sum_until_negative([1, 2, 3, -1, 4]))    
```

24. Write a function process_inputs that takes a list of strings. Use a for loop to print each string.
However, if the string "STOP" is encountered, use break to terminate the function execution
entirely

```python
def process_inputs(n):
    for i in n:
        if i == 'STOP':
            break
        print(i)

process_inputs(["Hello", "World", "STOP", "Python", "Code"])
```
