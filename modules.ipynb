```python
#GCD greatest common divisor

def gcd(x,y):
     while y != 0:
         x , y = y, x%y

     return abs(x)

gcd(12,18)
```

```python
#LCM (least commom Multiple)

def lcm(x,y):
    return abs(x*y) // gcd(x,y)

lcm(12,18)
```

```python
x = 10

def outer():
    y = 20
    def inner():
        z = 30
        print(y) 
        print(z)   # Local
        # Enclosing
        print(x)   # Global
        print(len([1,2,3]))  # Built-in
    inner()

outer()
```

```python
def factorial(n):
# Base case
    if n == 0 or n == 1:
        return 1
# Recursive case
    return n * factorial(n-1)


print(factorial(5))
```

```python
def divide(a, b):
    try:
        if b == 0:
            raise ZeroDivisionError("Division by zero not allowed")
        return a / b
    except ZeroDivisionError:
        print("Error: Division by zero")
    return None


print(divide(10, 2))
print(divide(10, 0))
```
