```python
list_var = [1, 2, 3, 4, 5]
tuple_var = (6, 7, 8, 9, 10)
set_var = {1, 2, 3, 4, 5}
dict_var = {'a': 1, 'b': 2, 'c': 3}

list_var.extend(tuple_var)

for item in set_var:
    if item not in list_var:
        list_var.append(item)

set_var = set(list_var)


# dict_var['d'] = min(set_var)
# dict_var['e'] = max(set_var)

dict_var.setdefault('d',min(set_var))
dict_var.setdefault('e',max(set_var))

print("list_var:",list_var)
print("set_var:",set_var)
print("dict_var:",dict_var)

```

```python
# Given variables
store1_sales = {"Laptop": 15, "Smartphone": 30, "Tablet": 10, "Headphones": 20}
store2_sales = {"Smartphone": 50, "Tablet": 40, "Camera": 25, "Headphones": 30}

# Write your code here
merged = {}

for product,sales in store1_sales.items():
    merged[product] = sales

for product,sales in store2_sales.items():
    if product in merged.keys():
        merged[product] += sales
    else:
        merged[product] = sales


max_sales = max(merged.values())
max_product = max(merged, key = merged.get)

print("Merged Sales Data:",merged)
print(f"Max Product: {max_product} with Total Sales: {max_sales}")
```
