```python
import numpy as np

# Given 2D NumPy array
data = np.array([[25, 180, 75], 
                 [30, 175, 68], 
                 [35, 170, 70], 
                 [40, 165, 80]])

# 1. Extract the first two rows and the first two columns
subarray = data[:2,:2]
print("Extracted 2x2 Subarray:")
print(subarray)

# 2. Extract the last row of the array
last_row = data[-1]
print(f"Last Row: {last_row}")

# 3. Extract all rows but only the second column
second_column = data[:,1]
print(f"Second Column: {second_column}")
```

```python
import numpy as np

# Weekly temperature data (C)
temperatures = np.array([
    25, 32, 29, 35, 28, 31, 30, 27, 33, 34, 26, 29, 30, 32, 33, 
    36, 28, 31, 30, 29, 35, 27, 28, 31, 34, 30, 29, 32, 35, 28, 
    33, 26, 29, 30, 34, 32, 36, 28, 31, 30, 29, 33, 28, 30, 31, 
    32, 34, 29, 27, 35, 31, 30, 28, 33, 34, 36, 29, 31, 30, 29, 
    35, 28, 32, 30, 27, 33, 34, 28, 29, 32, 31, 30, 27, 35, 29, 
    30, 32, 34, 31, 28, 33, 30, 29, 32, 36, 28, 35, 30, 29, 31, 
    34, 32, 33, 28, 30, 29
])

# Define the threshold
threshold = 30

# Mask indexing to filter temperatures
filtered_temperatures = temperatures[temperatures > threshold]

print(f"Filtered temperatures greater than {threshold}C:", filtered_temperatures)
```
