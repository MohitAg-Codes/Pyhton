43. Given a list of dictionaries representing employees [{'name': 'A', 'dept': 'HR'}, {'name': 'B',
'dept': 'IT'}, ...], write a program to group employees by department into a new dictionary

```python
employees = [
    {'name': 'A', 'dept': 'HR'},
    {'name': 'B', 'dept': 'IT'},
    {'name': 'C', 'dept': 'HR'},
    {'name': 'D', 'dept': 'IT'},
    {'name': 'E', 'dept': 'Finance'}
]

group = {}

for i in employees:
    dept = i['dept']
    name = i['name']

    if dept not in group:
        group[dept] = []
    group[dept].append(name)

print(group)
```

46. Generate the first n numbers of the Fibonacci sequence using a for loop and store them in a

```python
n = int(input())

a=0
b=1
feb = []

for i in range(n):
    feb.append(a)
    temp = a+b
    a = b
    b = temp

print(feb)
    
```
