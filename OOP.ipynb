```python
class Person:
    def __init__(self,name , age):
        self.name = name
        self.age = age

    def details(self):
        print(self.name, self.age)

p = Person('mohit',24)
p.details()
```

```python
#inheritence
class Animal:
    def __init__(self):
        print('animal')

class Dog(Animal):
    def __init__(self):
        super().__init__()
        print('dog')

d = Dog()
```

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

class Detail(Person):
    def __init__(self,name,age,gender):
        super().__init__(name,age)
        self.gender = gender

    def all(self):
        print(self.name, self.age, self.gender)

p = Detail('mohit',24,'male')
p.all()
```

```python
#override concept
class Animal:
    def sound(self):
        print('animal')

class Dog(Animal):
     def sound(self):
        print('dog')

d = Dog()
d.sound()
    
```

```python
#Encapsulation
# age --- public
# _age --- protected
# __age = private
class Person:
    def __init__(self):
        self.__age = 20  #private

    def set_age(self,age):
        self.__age = age

    def get_age(self):
        return self.__age

p = Person()
p.set_age(25)
print(p.get_age())
```

```python
#polymorphism

class Dog:
    def sound(self):
        print("Bark")


class Cat:
    def sound(self):
         print("Meow")  

animals = [Dog(), Cat()]

for a in animals:
    a.sound()
```

```python
#overloading concept
class Calculator:
    def add(self, a, b, c=0):
        return a + b + c

calc = Calculator()

print(calc.add(2, 3))        # 2 arguments
print(calc.add(2, 3, 4))     # 3 arguments
```

```python
#abstract concept
from abc import ABC , abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print('dog')

d = Dog()
d.sound()
```
