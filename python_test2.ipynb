```python
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2,n):
        if n%i == 0:
            return False

    return True
    

nums = list(map(int, input().split()))

for num in nums:
    
    if is_prime(num):
        print(num, end=" ")
```

```python
def power(x, n) :
    if n == 0:
        return 1

    return x * power(x, n-1)

x = int(input())
n = int(input())

# Output
print(power(x, n))
```

```python
def validate_user(age, salary):
    if not (18 <= age <= 100):
        raise ValueError("Age must be between 18 and 100")
    if salary <= 0:
        raise ValueError("Salary must be positive")
    if salary > 10**4:  
        raise ValueError("Salary exceeds the maximum allowed value")
    return "Both age and salary are valid"


# Input from user
try:
    age = int(input())
    salary = int(input())
    result = validate_user(age, salary)
    print(result)
except ValueError as e:
    print(e)
```

```python
from datetime import datetime

def days_between_dates(date1, date2):
    
    d1 = datetime.strptime(date1, "%Y-%m-%d")
    d2 = datetime.strptime(date2, "%Y-%m-%d")
    
   
    delta = d2 - d1
    
    
    return delta.days


date1 = input()
date2 = input()


print(days_between_dates(date1, date2))
```

```python
# List of operations with empty lists
operations = [
    ["Assembly", []],
    ["Testing", []],
    ["Packaging", []],
    ["Shipping", []]
]

# Define the functions
def assembly(operations):
    operations[0][1].append("Assembling parts")
    operations[0][1].append("Fixing components")

def testing(operations):
    operations[1][1].append("Running diagnostics")
    operations[1][1].append("Checking quality")

def packaging(operations):
    operations[2][1].append("Wrapping products")
    operations[2][1].append("Labeling packages")

def shipping(operations):
    operations[3][1].append("Arranging transportation")
    operations[3][1].append("Updating delivery details")


assembly(operations)
testing(operations)
packaging(operations)
shipping(operations)

# Display the results
for op_name, op_list in operations:
    print(f"{op_name}: {op_list}")
```

```python
import math
def prime_factors(n: int) -> list:
    factors = []

    # Divide out 2 first
    while n % 2 == 0:
        factors.append(2)
        n //= 2

    # Check odd numbers up to sqrt(n)
    for i in range(3, int(math.isqrt(n)) + 1, 2):
        while n % i == 0:
            factors.append(i)
            n //= i

    # If remaining n is prime
    if n > 2:
        factors.append(n)

    return factors

# Example usage
n = int(input())
print(prime_factors(n))
```
