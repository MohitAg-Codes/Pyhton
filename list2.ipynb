```python


a = [1, [2, 3]]
b = a.copy()   # shallow copy

b[0] = 10          # changes only in b
b[1][0] = 20       # changes reflect in both
print(a)           # [1, [20, 3]]
print(b)           # [10, [20, 3]]
```

```python
import copy

a = [1, [2, 3]]
b = copy.deepcopy(a)   # deep copy

b[1][0] = 20

print(a)   # [1, [2, 3]]
print(b)   # [1, [20, 3]]
```
