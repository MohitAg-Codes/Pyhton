```python
confusion_matrix = [
    [50, 10],  # Row 1: TP, FN
    [5, 35]    # Row 2: FP, TN
]

# Write your code here
TP =  confusion_matrix[0][0]
FN =  confusion_matrix[0][1]
FP =  confusion_matrix[1][0]

precision = TP / (TP + FP)

recall = TP / (TP + FN)

F1_score = 2*((precision * recall)/(precision + recall))

print('Precision:',round(precision,2))
print('Recall:',round(recall,2))
print('F1_score:',round(F1_score,2))
```

```python
def calculator(operation, *args):
    pass
    # 1. Add : 10 + 2 + 3 + 14 = 29
    # 2. Subtract : 10 - 2 - 3 - 5 = 0
    # 3. Multiply : 3 * 2 * 2 = 12
    def add():
        return sum(args)

    def multiply():
        result = 1
        for i in args:
            result *= i
        return result

    def subtract():
        result = args[0]
        for i in args[1:]:
            result -= i
        return result
        

    if operation == 'add':
        return add()
    elif operation == 'multiply':
        return multiply()
    elif operation == 'subtract':
        return subtract()
    else:
        return 'Invalid operation'

print(calculator('subtract',10,20,30))
```

```python
data = [15, 20, 35, 40, 50, 45, 60, 55]

def  median(data):
    n = len(data)
    mid = n//2

    if n %2 == 1:
        median = data[mid]
    else:
        median = (data[mid] + data[mid - 1]) / 2

    return median


def calculate_iqr(data):
    data.sort()

    n = len(data)
    mid = n // 2

    if n % 2 == 1:
        lower = data[: mid]
        upper = data[mid+1 : ]
    else:
        lower = data[:mid]
        upper = data[mid:]


    q1 = median(lower)
    q3 = median(upper)

    iqr = q3 - q1

    return iqr

print(calculate_iqr(data))



    
```

```python
cart_items1 = {"apple": (3, 5), "milk": (2, 2), "bread": (1.5, 3)}
discounts1 = {"apple": 10, "milk": 5}
delivery_charge1 = 3

calculate_bill(cart_items1, discounts1, delivery_charge1)


def calculate_bill(cart_items, discounts, delivery_charge=0):
    print("---- Bill Details ----")

    bill = 0
    for key,(price,quantity) in cart_items.items():
        total = price * quantity
        discount = discounts.get(key,0)
        amount = total * discount / 100
        final = total - amount

        bill = bill + final

        print(f'{key.capitalize()} - {quantity} x ${price} - {discount}% discount = ${final}')

    print(f'Delivery Charge: ${delivery_charge}')

    bill = bill + delivery_charge
    print(f'Total Bill: ${bill}')

return bill
```

```python
data = [
    ('Aarav', 25, 'Delhi'),
    ('Vivaan', 30, 'Mumbai'),
    ('Reyansh', 20, 'Bangalore'),
    ('Aditya', 35, 'Chennai'),
    ('Kabir', 28, 'Delhi'),
    ('Vihaan', 22, 'Mumbai')
]

def filter_and_sort(data, min_age=None, city=None):
    filtered_data = [
        person for person in data 
        if (min_age is None or person[1] >= min_age) and (city is None or person[2] == city)
    ]
    
    sorted_data = sorted(filtered_data, key=lambda x: (x[2], x[1]))
    
    return sorted_data

result = filter_and_sort(data, min_age=25)

print(result)
```

```python
def recursive_function(n):
    if n <= 1:
        return n
    else:
        return recursive_function(n - 1) + recursive_function(n - 2)

result = recursive_function(6)
print(result)
```
